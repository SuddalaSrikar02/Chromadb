```python
emb=HuggingFaceEmbeddings(model="all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import Chroma
```

```python
# create a database
```

```python
from langchain_core.documents import Document
```

```python
data=["asdsad","Asdsadsa","asdsad"]
```

```python
data=[Document(page_content="langchain is a framework to create llm based application",metadata={"topic":"langchain","difficult":"easy"}),
      Document(page_content="langchain contains four important coponents langchain,langgrpah,langsmith,lagserve",metadata={"topic":"langchain","difficult":"medium"}),
      Document(page_content="ml stands for machine learning",metadata={"topic":"ml","difficult":"easy"}),
      Document(page_content="ml can be divided into four learning ",metadata={"topic":"ml","difficult":"easy"})]
```

```python
["doc{}".format(y) for y in range(len(data))]
```

```python
# create
db=Chroma.from_documents(documents=data,embedding=emb,persist_directory="db3",ids=["doc{}".format(y) for y in range(len(data))])
db.persist()
```
